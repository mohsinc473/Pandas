```python
dmap = {
    "AI" : 0,
    "BC" : 1,
    "IOT" : 2,
    "CNC" : 3,
    np.nan : 4,
}

data1.Course.map(dmap)
```

```python
dic = {}

for i,v in enumerate(data1.Course.unique()):
    dic[v] = i

dic
```

```python
data1["Updated Courses"] = data1.Course.map(dic)
data1
```

```python
def updateFee(c):
    if c == "AI":
        return 7000
    elif c == "BC":
        return 4000
    elif c == "IOT":
        return 4500
    elif c == "CNC":
        return 4200
    else:
        return 0

data1["Updated Fee"] = data1.Course.apply(updateFee) 
data1
```

```python
df2 = pd.DataFrame(np.random.randint(1,100, size=(6,5)),columns=['A','B','C','D','E'])
df2
```

```python
pd.concat([data1,df2],axis=0)
```

```python
df3 = pd.concat([data1,df2],axis=1)
df3
```

```python
df3["Total"] = df3.A + df3.B + df3.C + df3.D + df3.E 
df3["Percentage"] = (df3.A + df3.B + df3.C + df3.D + df3.E)/500*100 

def grade(p):
    if p>=90:
        return "A"
    elif p>=80:
        return "B+"
    elif p>=70:
        return "B"
    elif p>60:
        return "C+"
    elif p>50:
        return "C"
    else:
        return "Fail"

df3["Grade"] = df3.Percentage.apply(grade)
df3
        
```

```python
def user_fn(course,c1,c2,c3,c4,c5):
    if course == "AI":
        return np.array([c1,c2,c3,c4,c5]).sum()
    elif course == "IOT":
        return np.array([c1,c2,c3,c4,c5]).prod()
    elif course == "CNC":
        return sum(np.array([c1,c2,c3,c4,c5]).cumsum())
    else:
        return "0"

df3["Validation"] = df3[["Course","A","B","C","D","E"]].apply(lambda x:user_fn(*x),axis=1)
df3
```
